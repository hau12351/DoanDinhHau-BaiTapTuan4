```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import KBinsDiscretizer, StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from mlxtend.frequent_patterns import apriori, association_rules

try:
    df = pd.read_csv('HeartDiseaseTrain-Test.csv')
    df.columns = df.columns.str.strip()
    print(" c file thnh cng!")
except FileNotFoundError:
    print("LI: Bn cha upload file 'HeartDiseaseTrain-Test.csv' ln Colab!")

cat_cols = ['sex', 'chest_pain_type', 'fasting_blood_sugar', 'rest_ecg',
            'exercise_induced_angina', 'slope', 'vessels_colored_by_flourosopy',
            'thalassemia', 'target']
cat_cols = [c for c in cat_cols if c in df.columns]
num_cols = [c for c in df.columns if c not in cat_cols]

for c in num_cols: df[c] = df[c].fillna(df[c].median())
for c in cat_cols: df[c] = df[c].fillna(df[c].mode()[0])

disc = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='quantile')
num_disc = disc.fit_transform(df[num_cols])
num_disc_df = pd.DataFrame(num_disc, columns=[c + '_bin' for c in num_cols])

def label_bins(x):
    return {0.0: 'low', 1.0: 'mid', 2.0: 'high'}.get(float(x), x)

for c in num_disc_df.columns:
    num_disc_df[c] = num_disc_df[c].apply(label_bins)

apriori_input = pd.concat([df[cat_cols], num_disc_df], axis=1)
apriori_df = pd.get_dummies(apriori_input.astype(str), dtype=bool)

freq_items = apriori(apriori_df, min_support=0.1, use_colnames=True)
rules = association_rules(freq_items, metric='confidence', min_threshold=0.6)
print(f"S lut tm c: {len(rules)}")
display(rules.sort_values('lift', ascending=False).head(10))

X_scaled = StandardScaler().fit_transform(df[num_cols])
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(X_scaled)
print(f"Silhouette Score: {silhouette_score(X_scaled, df['cluster'])}")
```
